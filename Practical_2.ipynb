```python
import numpy as np
from scipy.stats import norm

# Sample dataset
data = np.array([2.3, 2.5, 2.8, 3.0, 3.2])

# Apply MLE to estimate mean and standard deviation
mu, std = norm.fit(data)

print("Estimated Mean (mu):", mu)
print("Estimated Standard Deviation (sigma):", std)
```
