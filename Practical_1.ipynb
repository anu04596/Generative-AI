```python
import numpy as np
import matplotlib.pyplot as plt

# Generate Normal Distribution (mean=0, std=1)
normal_data = np.random.normal(loc=0, scale=1, size=1000)

# Generate Uniform Distribution (range 0 to 1)
uniform_data = np.random.uniform(low=0, high=1, size=1000)

# Plot both distributions
plt.hist(normal_data, bins=30, alpha=0.5, label='Normal Distribution')
plt.hist(uniform_data, bins=30, alpha=0.5, label='Uniform Distribution')

plt.legend()
plt.title("Random Samples from Distributions")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```
