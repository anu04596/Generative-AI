```python
# Step 1: Import libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input
import numpy as np

# Step 2: Better text dataset (IMPORTANT)
text =  """
machine learning is a field of artificial intelligence
that focuses on building systems that learn from data
deep learning is a subset of machine learning
neural networks are used for pattern recognition
artificial intelligence is transforming industries
data science involves statistics and programming
"""

# Step 3: Create character mappings
chars = sorted(list(set(text)))
char_to_idx = {c: i for i, c in enumerate(chars)}
idx_to_char = {i: c for i, c in enumerate(chars)}

# Step 4: Prepare sequences
sequence_length = 10
X = []
y = []

for i in range(len(text) - sequence_length):
    seq = text[i:i + sequence_length]
    target = text[i + sequence_length]

    X.append([char_to_idx[c] for c in seq])
    y.append(char_to_idx[target])

X = np.array(X)
y = np.array(y)

# Step 5: Normalize input
X = X / float(len(chars))

# Reshape for RNN [samples, timesteps, features]
X = np.reshape(X, (X.shape[0], X.shape[1], 1))

# Step 6: Build model (NO WARNING)
model = Sequential([
    Input(shape=(sequence_length, 1)),
    LSTM(128),
    Dense(len(chars), activation='softmax')
])

# Step 7: Compile
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam')

# Step 8: Train (increase epochs for better output)
model.fit(X, y, epochs=200, batch_size=16, verbose=1)

# Step 9: Generate text (IMPROVED SAMPLING)
start = X[np.random.randint(0, len(X))]
generated = ""

for i in range(150):
    pred = model.predict(start.reshape(1, sequence_length, 1), verbose=0)

    # Probabilistic sampling (better than argmax)
    pred = pred[0]
    pred = np.log(pred + 1e-8)
    pred = np.exp(pred) / np.sum(np.exp(pred))

    idx = np.random.choice(len(pred), p=pred)
    generated += idx_to_char[idx]

    # Update sequence
    start = np.append(start[1:], idx / float(len(chars)))

print("\nGenerated Text:\n")
print(generated)
```
